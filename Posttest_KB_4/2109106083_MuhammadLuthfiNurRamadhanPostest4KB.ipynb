### Data Cleaning

```python
df.info()
```

#### Nilai Null

```python
#mengecek data null
print("Jumlah record yang memiliki nilai null:")
print(df.isna().sum())
```

```python
imputer_mean = SimpleImputer(strategy = 'mean')
imputer_median = SimpleImputer(strategy = 'median')
imputer_most_frequent = SimpleImputer(strategy = 'most_frequent')
```

- Mengganti nilai null dengan tipe data float dengan mean

```python
df['critics_vote'] = imputer_mean.fit_transform(df[['critics_vote']])
```

```python
#mengecek data null
print("Jumlah record yang memiliki nilai null:")
print(df.isna().sum())
```

- mengubah type data total vote dari float menjadi int

```python
df["total_votes"] = df["total_votes"].astype("int64")
df
```

- Mengganti nilai null dengan tipe data float dengan median

```python
df['total_votes'] = imputer_median.fit_transform(df[['total_votes']])
```

```python
#mengecek data null
print("Jumlah record yang memiliki nilai null:")
print(df.isna().sum())
```

- Mengganti nilai null dengan tipe data object dengan modus

```python
df['genre'] = imputer_most_frequent.fit_transform(df[['genre']])
```

```python
#mengecek data null
print("Jumlah record yang memiliki nilai null:")
print(df.isna().sum())
```
